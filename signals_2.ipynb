# 1.2.6

```python
import matplotlib.pyplot as plt
import numpy as np

n_h = np.array([0, 1, 2, 3, 4])
h = np.array([1, 2, 1, 0, -2])

n_x = np.array([0, 1])
x = np.array([1, 0.5])

n_y = np.convolve(n_x, n_h)
y = np.convolve(x, h)

plt.figure(figsize=(14, 4))
plt.subplot(1, 3, 1)
plt.stem(n_x, x, basefmt=" ")
plt.title('Input Signal x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.grid(True)

plt.subplot(1, 3, 2)
plt.stem(n_h, h, basefmt=" ")
plt.title('Impulse Response h[n]')
plt.xlabel('n')
plt.ylabel('h[n]')
plt.grid(True)

plt.subplot(1, 3, 3)
plt.stem(range(len(y)), y, basefmt=" ")
plt.title('Output Signal y[n]')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.grid(True)

plt.tight_layout()

plt.show()

n_y, y
```

# 1.2.10

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(-5, 10, 1)

u = np.heaviside(n, 1)

y = 2 * u - np.heaviside(n-4, 1)

plt.stem(n, y, use_line_collection=True)
plt.xlabel('n')
plt.ylabel('y(n)')
plt.title('Output of the System y(n) = 2u(n) - u(n-4)')
plt.grid(True)
plt.show()
```
